Import Data

```python
!pip install requests==2.31.0
import pandas as pd
import requests
import io

url = "https://raw.githubusercontent.com/Gani077/YBI_ML_Project/main/MPG.csv"

response = requests.get(url)
response.raise_for_status()

try:
    df = pd.read_csv(io.StringIO(response.text))
except pd.errors.ParserError as e:
    print(f"Error parsing CSV: {e}")
    print("Trying with different delimiters")
    try:
      df = pd.read_csv(io.StringIO(response.text), sep='\s+')
    except pd.errors.ParserError as e:
      print(f"Error parsing CSV: {e}")
      print("Please inspect the file at", url, "for errors.")

print(df.head())
```

Describe Data

```python
df.head()
```

Data Preprocessing

```python
df.info()
```

```python
df.describe()
```

removing missing values

```python
df=df.dropna()
df.info()
```

Data Visualization

```python
sns.pairplot(df,x_vars=['displacement','horsepower','weight','acceleration','mpg'], y_vars=['mpg'])
```

```python
sns.regplot(x='displacement',y='mpg',data=df);
```

Define Target Variable (Y) and Feature Variables (X)

```python
df.columns
```

```python
y=df['mpg']
y.shape
```

```python
X=df[['displacement','horsepower','weight','acceleration']]
```

```python
X.shape
```

```python
X
```

SCALING DATA

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
X
```

```python
pd.DataFrame(X).describe()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Modeling-Linear regression model

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

Prediction

```python
y_pred=lr.predict(X_test)
y_pred
```

Model Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
